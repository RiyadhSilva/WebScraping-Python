```python
import itertools
import requests
import time
from bs4 import BeautifulSoup
```

```python
# Definindo uma função para efetuar o download da página
def download(url, num_retries=2):
    print('Dowloading: ', url)
    page = None
    try:
        response = requests.get(url)
        page = response.text
        if response.status_code >= 400:
            print('Download error: ', response.text)
            if num_retries and 500 <= response.status_code < 600:
                return dowload(url, num_retries - 1)
    except requests.exceptions.RequestException as e:
            print('Download error: ', e.reason)
    return page
```

```python
lst_soups = []
def crawl_site(url):
    for page in itertools.count(1):
        pg_url = '{}{}'.format(url, page)
        html = download(pg_url)
        soup = BeautifulSoup(html, 'html.parser')
        if(soup.contents[0]=='Invalid record' or html is None):
            break
        else:
            lst_soups.append(soup)
            time.sleep(3)
```

```python
crawl_site('http://example.webscraping.com/places/default/view/-')
```

```python
file = open('country_pages_soups.txt', 'w')
file.write(str(lst_soups))
file.close()
```

```python
area_global = 0
population_global = 0
for soup in lst_soups:
        area = soup.find('table')
        area = soup.find('table').find(id='places_area__row').find(class_='w2p_fw').contents
        area_int = int(area[0].replace(',','').replace('square kilometres', '').replace(' ', ''))
        area_global += area_int
        population = soup.find('table').find(id='places_population__row').find(class_='w2p_fw').contents
        population_int = int(population[0].replace(',',''))
        population_global += population_int
        print('%d %d'%(population_global, area_global))
print('A densidade populacional é de %d hab/km²'%((population_global/area_global)))
```
