```python
import pandas as pd
from bs4 import BeautifulSoup
import matplotlib as plot
import numpy as np
```

```python
file = open('country_pages_soups.txt', 'r')
```

```python
lst = file.read().split('</html>')
```

```python
lst.remove(lst[len(lst)-1])
```

```python
soup = BeautifulSoup(lst[0], 'html.parser')
```

```python
soup.find('table').find(id='places_country__row').find(class_='w2p_fw').contents
```

```python
nome = []
populacao   = []
area        = []
for page in lst:
    soup = BeautifulSoup(page, 'html.parser')
    n = soup.find('table').find(id='places_country__row').find(class_='w2p_fw').contents
    nome.append(n[0])
    p = soup.find('table').find(id='places_population__row').find(class_='w2p_fw').contents
    populacao.append(p[0].replace(',',''))
    a = soup.find('table').find(id='places_area__row').find(class_='w2p_fw').contents
    area.append(a[0].replace(',','').replace('square kilometres',''))
```

```python
mundo = {
    'nome': nome,
    'populacao': populacao,
    'area - km²': area
}
```

```python
df = pd.DataFrame(mundo, columns=['nome', 'area - km²', 'populacao'])
df
```

```python
df.dtypes
```

```python
df['populacao'] = df['populacao'].astype(np.int64)
df['area - km²'] = df['area - km²'].astype(np.int64)
```

```python
df.dtypes
```

```python
# Alterando índices
df.index = df['nome']
df
```

```python
# Apagando colunas duplicadas
del df['nome']
df
```

```python
df.describe()
```

```python
# Criando uma nova coluna pegando os dados de duas colunas distintas
df['densidade demografica'] = df['populacao'] / df['area - km²']
```

```python
df
```

```python
# Tratando valores nulos
df.fillna('-', inplace=True)
df
```

```python
writer = pd.ExcelWriter('dados.xlsx')
df.to_excel(writer, 'Sheet1', float_format='%.3f')
```

```python
writer.save()
```
