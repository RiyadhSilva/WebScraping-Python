```python
# Importando pacote pandas externo ao Python como 'pd'
import pandas as pd
```

```python
# Lendo página html
page = pd.read_html('http://www.imdb.com/chart/boxoffice')
```

```python
page
```

```python
# Imprimindo os dados da primeira tabela do site
page[0]
```

```python
# Criando um dataFrame para armazenar os valores
dataFrame = page[0]
```

```python
# 'Dropando' colunas de conteúdo vazio
dataFrame = dataFrame.drop(['Unnamed: 0', 'Unnamed: 5'], axis=1)
dataFrame
```

```python
# Renomeando o nome das colunas
dataFrame = dataFrame.rename(index=str, columns={'Title':'Título', 'Weekend':'Fim de semana', 'Gross':'Bruto', 'Weeks':'Semanas'})
dataFrame
```

```python
# Armazenando localmente no formato csv
dataFrame.to_csv('questao_02.csv')
```

```python
# Filtrando os itens de uma coluna específica
valores = dataFrame.filter(items=['Fim de semana'])
valores
```

```python
# Pegando apenas os números (Sem '$' e sem 'M')
lst_valores = []
for lista in valores.values:
    for valor in lista:
        lst_valores.append(valor[1:(len(valor)-1)])
```

```python
# Imprimindo a lista de valores
lst_valores
```

```python
# Convertendo os itens da lista anterior em objetos do tipo float e adicionando-os em uma nova lista
lst_valores_ft = []
for valor in lst_valores:
    lst_valores_ft.append(float(valor))
```

```python
# Imprimindo a nova lista
lst_valores_ft
```

```python
print('A arrecadação semanal foi de ' + str(sum(lst_valores_ft)) + 'M')
```
