```python
# Importando pacote request externo ao Python
import requests
```

```python
# Definindo uma função para efetuar o download da página
def download(url, num_retries=2):
    print('Downloading: ', url)
    page = None
    try:
        response = requests.get(url)
        page = response.text
        if response.status_code >= 400:
            print('Download error:', response.text)
            if num_retries and 500 <= response.status_code < 600:
                return download(url, num_retries - 1)
    except requests.exceptions.RequestExceptions as e:
        print('Download error: ', e.reason)
    return page
```

```python
# Importanto pacote externo ao Python, efetuando o download do site e iniciando a 'sopa'
from bs4 import BeautifulSoup
url = 'http://www.imdb.com/chart/boxoffice'
html = download(url)
soup = BeautifulSoup(html, 'html.parser')
```

```python
# Capturando a tag tabela dentro do site e armazenando os dados em uma variável chamada t
t = soup.find('table')
```

```python
# Imprimindo a variável
t
```

```python
# Imprimindo o conteúdo da variável
t.contents
```

```python
print(t.find_all(class_ = 'titleColumn')[3].get_text())
print(t.find_all(class_ = 'ratingColumn')[6].get_text())
print(t.find_all(class_ = 'secondaryInfo')[3].get_text())
print(t.find_all(class_ = 'weeksColumn')[3].get_text())
```

```python
qtd = len(t.find_all(class_ = 'titleColumn'))
```

```python
position = 0
list_dict = []   
```

```python
while position < qtd:
    titulo     = t.find_all(class_ = 'titleColumn')[position].get_text().replace('\n','')
    fim_de_semana = t.find_all(class_ = 'ratingColumn')[position*2].get_text().replace(' ', '').replace('\n','')
    total = t.find_all(class_ = 'secondaryInfo')[position].get_text()
    semanas = t.find_all(class_ = 'weeksColumn')[position].get_text()
    filme_dic = {'Título': titulo, 'Fim de semana': fim_de_semana, 'Total': total, 'Semanas': semanas}
    list_dict.append(filme_dic)
    position+=1
```

```python
list_dict
```

```python
# Importa pacote externo ao Python
import pandas as pd
```

```python
# Cria uma tabela recebendo a lista de dicionários como parâmetro e a ordem das colunas passa a ser 'Nome' e 'Avaliação'
tabela = pd.DataFrame(list_dict).filter(items=['Título','Fim de semana', 'Total', 'Semanas'])
```
